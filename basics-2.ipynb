```python
#Line graph

w=[1,2,4,5]
d=[1,7,9,1]
plt.title('Line graph')
plt.xlabel('input')
plt.ylabel('output')
plt.plot(w,d,c='orange',marker='D') #C represents color #D represents diamond
```

```python
#What ever task LISTS do, the same task is done by NUMPY in a faster way
z=np.array([1,2,4,5])
u=np.array([1,7,9,1])
plt.title('Line graph')
plt.xlabel('input')
plt.ylabel('output')
plt.plot(z,u,c='orange',marker='D')
```

```python
y=np.array([5,3,8,9])
plt.plot(y,linewidth=10)
```

```python
#Cricket analysis

overs=np.array([0,5,10,15,20])
csk=np.array([10,60,110,160,210])
mi=np.array([5,15,60,80,150])
plt.title('IPL',fontsize=20)
plt.xlabel('OVERS')
plt.ylabel('SCORE-WICKETS')
plt.plot(overs,csk,c='y',marker='D',label='CSK')
plt.plot(overs,mi,c='b',marker='D',label='MI')
plt.legend() # To display the line label indicator
```

```python
x=np.array([5,10,15,20])
y=[1,2,3,4]
plt.title('Watch data',fontsize=20)
plt.plot(x,y)
plt.grid()
```

```python
#SUB PLOT
fig = plt.figure()
x1 = [1,2,3,4]
x2 = [5,6,7,8]
y1 = np.random.randint(0,10,4)
y2 = np.random.randint (0,10,4)
ax1 = fig.add_subplot (221)
ax2 = fig.add_subplot (224)
ax1.plot(x1, y1)
ax2.plot(x2,y2)
```
